```python
print("Punto 1")
print("")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("Punto 2")
data= "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
data= pd.read_csv(data)
```

```python
print("Punto 3")
print("")
print(data[-5:])
print("")
data.info()
print("")
print ("Dimensiones", data.shape)
print("")
data.describe()
```

```python
print("Punto 4")
print(data.isnull().sum())
```

```python
print("Punto 5")
data["volumen_aprox"] = data["x"]*data["y"]*data["z"]
data.head()
```

```python
print("Parte 2: Visualización de datos")
plt.figure()

plt.boxplot(data["price"])

plt.title("Gráfico de caja")
plt.xlabel("Price")
plt.ylabel("Valor")

plt.show()

print("")
x = data["volumen_aprox"]
y = data["price"]
plt.scatter(x, y, color='red', label="")
plt.title("Punto 2: Gráfico de dispersión")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

ideal = len(data[data["cut"] == "Ideal"])
premium = len(data[data["cut"] == "Premium"])
very_good = len(data[data["cut"] == "Very Good"])
good = len(data[data["cut"] == "Good"])
fair = len(data[data["cut"] == "Fair"])

tipos = ["Ideal", "Premium", "Very Good", "Good", "Fair"]
cantidades = [ideal, premium, very_good, good, fair]

plt.bar(tipos, cantidades)

plt.title("Cantidad de Piezas por Tipo de Corte")
plt.xlabel("Tipo de Corte")
plt.ylabel("Cantidad")

plt.show()
```
